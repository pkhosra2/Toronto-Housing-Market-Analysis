### Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
!pip install seaborn
```

```python
import seaborn as sns
```

### Reading Our DataFrame


```python
df = pd.read_csv('ward_housing_income.csv') #import csv file into a dataframe
df.head() #print dataframe
```

### Print Heatmap Visualization

```python
#changing our dataset using the pivot table
result = df.pivot_table(columns=['Income'], index=['Ward'], values=['Count'])
#setting our figure size
plt.figure(figsize=(20,20))
#creating our heatmap using seaborn
sns.heatmap(result, annot=False, fmt="g", cmap="Blues")
#print heatmap title
plt.title('Toronto Household Income Distribution by Ward',fontsize=20)
plt.show()
```
