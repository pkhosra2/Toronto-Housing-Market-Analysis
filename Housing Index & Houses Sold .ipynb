# Toronto Housing Index & Houses Sold Visualization

### Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Housing Index & Houses Sold DataFrame


```python
df = pd.read_csv('housing_index.csv',index_col = 'Month') #read the housing index csv file, setting the inde as the month
df.head() #print the dataframe
```

### Parsing The Original DataFrame Into Seperate DataFrames


For the first dataframe we drop the Sales Count column from the original dataframe to isolate the housing index value

```python
df_index = df.drop(["Sales Count"],axis=1) #setting the new dataframe by dropping the 'Sales Count' column
df_index.head() #Printing the new dataframe 
```

For the second dataframe we drop the Index column from the original dataframe to isolate the Sales Count value


```python
df_sales = df.drop(['Index'],axis=1) #setting the new dataframe by dropping the 'Housing Index Value' column
df_sales.head() #Printing the new dataframe
```

### Plotting Both Parsed Line Graphs

```python
plt.rcParams["figure.figsize"] = (20,10) #set the size of both figures

index = df_index.plot(kind='line', color='red'); #plotting the housing index line graph
plt.ylabel('Index Value') #set y axis
plt.xlabel('Month of Year') #set x axis
plt.title('Toronto Housing Index Value') #set title
plt.grid(True)

sales = df_sales.plot(kind='line'); #plotting the sales count line graph
plt.ylabel('# Toronto House Sales') #set y axis
plt.xlabel('Month of Year') #set x axis
plt.title('Toronto Housing Sales Value') #set title
plt.grid(True)

loc, labels = plt.xticks() #set both plot legends
index.set_xticklabels(labels, rotation=45)
sales.set_xticklabels(labels, rotation=45)
```
