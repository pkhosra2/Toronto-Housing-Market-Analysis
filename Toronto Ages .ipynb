### Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
!pip install seaborn
```

```python
import seaborn as sns
```

### Importing Our DataFrame

```python
df = pd.read_csv('Toronto_Ages.csv', index_col = 'Age Groups')
df.head()
```

### Clean Up Our Data 

```python
df = df.drop(df.index[[0,4,15]])
```

```python

# y-axis in bold
rc('font', weight='bold')

male = df['Male']
female = df['Female']
```

### Plotting Our Visualization

```python

colors = {'#52ABE6','#FCA5CD'}
plt.rcParams["figure.figsize"] = (20,20)
ages = df.plot(kind='bar', stacked=True, color=colors, edgecolor="white"); #stacked

loc, labels = plt.xticks()
ages.set_xticklabels(labels, rotation=45)
plt.ylabel('Total')
plt.xlabel('Age Group')
plt.title('Age Spread of Toronto in 2016')

plt.show()
```
